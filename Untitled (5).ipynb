```python
model=Sequential([
    tf.keras.Input(shape=(x_train.shape[1])),
    Dense(256,activation='relu',name='2',kernel_regularizer=tf.keras.regularizers.L2(0.001)),
    Dense(128,activation='relu',name='1',kernel_regularizer=tf.keras.regularizers.L2(0.001)),
    Dense(64,activation='relu',name='61',kernel_regularizer=tf.keras.regularizers.L2(0.001)),
    Dense(32,activation='relu',name='62',kernel_regularizer=tf.keras.regularizers.L2(0.01)),
    Dense(16,activation='relu',name='623',kernel_regularizer=tf.keras.regularizers.L2(0.01)),
    Dense(1,activation='sigmoid',name='7')
])
```

```python
class StopAtAccuracy(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if logs.get('val_accuracy') is not None and logs.get('val_accuracy') >= 0.89:
            self.model.stop_training = True
```

```python
model.compile(
    loss=BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(0.0001),
     metrics=['accuracy', 'AUC']
)
his=model.fit(
    x_train,y_train,
    epochs=250,
    batch_size=6,
    validation_data=(x_test,y_test),
    callbacks=[StopAtAccuracy()]
)
```

```python
model.evaluate(x_test,y_test)
```
